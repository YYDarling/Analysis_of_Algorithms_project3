**MainCode : Algorithm Design**

```python
def algoriam_design_1(num_of_nodes, num_of_distinct_labels):

  # list_of_node = [[1, 2], [0, 3, 4], [0, 5, 6], [1, 7, 8, 9], [1, 10, 11], [2], [2], [3], [3], [3], [4], [4]]
  
  # Initialize variables
  len_node = len(num_of_nodes)
  graph = (len_node, num_of_distinct_labels)
  full_len_node = len_node + 1
  whole_mat = np.full(graph, full_len_node)
  ans = [-1]*len_node

  # Label node 0
  labeled = set()
  cur_label = np.argmax(whole_mat[0])
  ans[0] = cur_label
  algoriam_design_2(labeled, 0, 0, cur_label, whole_mat, num_of_nodes)

  # Label remaining nodes
  for i in range(0, len_node):
      neighbors = num_of_nodes[i]
      for node in neighbors:
          if ans[node] == -1:
              labeled = set()
              cur_label = np.argmax(whole_mat[node])
              ans[node] = cur_label
              algoriam_design_2(labeled, node, 0, cur_label, whole_mat, num_of_nodes)

  return ans, whole_mat


def algoriam_design_2(labeled, i, distance, cur_label, whole_mat, num_of_nodes):
    # Update node i
    if i in labeled:
        return
    whole_arr = whole_mat[i]
    new_dis = min(whole_arr[cur_label], distance)
    whole_mat[i][cur_label] = new_dis
    labeled.add(i)

    # Update neighbors
    neighbors = num_of_nodes[i]
    for j in neighbors:
        algoriam_design_2(labeled, j, distance+1, cur_label, whole_mat, num_of_nodes)
```
